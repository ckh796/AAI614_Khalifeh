```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import ssl


ssl_context = ssl._create_unverified_context()

url = "https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.html"

# Fetch the webpage content
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')

# Find the table with id "t2"
table = soup.find('table', {'id': 't2'})

# Extract table headers
headers = [th.get_text(strip=True) for th in table.find_all('th')]

# Extract table rows
rows = []
for tr in table.find_all('tr'):
    cells = [td.get_text(strip=True) for td in tr.find_all(['td', 'th'])]
    if len(cells) == len(headers):  # Ensure row matches header length
        rows.append(cells)

# Create a DataFrame
df = pd.DataFrame(rows, columns=headers)
df.head()
```
